# Problem 1

```python
import os
import glob
import shutil
import datetime

os.mkdir("/content/all_images")
source_dir=r"/content/drive/MyDrive/dairies"

files = glob.glob(source_dir + r'/**/*.jpg', recursive=True)

# Move all images into one folder
for file_name in files:
    shutil.copy(os.path.join(source_dir, file_name), "/content/all_images")
```

```python
image_details = []
for path in files:
    img = {'Image': os.path.basename(path).split('_')[1],
           'Image Size': os.stat(path).st_size,
           'Image Modification Date': datetime.datetime.fromtimestamp(os.stat(path).st_mtime)}
    image_details.append(img)
```

```python
# Creating a csv report with the following:
'''
→ image name “with the prefix discarded.”
→ image size
→ image last modification date
'''

import csv

keys = image_details[0].keys()

with open('ImagesReport.csv', 'a', newline='') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(image_details)
```

# Problem 2

```python
# Converting txt to JSON file
import json

with open("image1.txt", "r") as file:
    lines = file.readlines()

data = []
for line in lines:
    values = line.strip().split()
    annotations = {
        "image rotation": 0,
        "value": {
            "x": float(values[1]),
            "y": float(values[2]),
            "width": float(values[3]),
            "height": float(values[4]),
            "rotation": 0,
            "rectangleLabels": ["object"]
        }
    }
    data.append(annotations)

# Save JSON data to a file
with open("image1.json", "w") as Jsonf:
    json.dump(data, Jsonf, indent=4)
```
